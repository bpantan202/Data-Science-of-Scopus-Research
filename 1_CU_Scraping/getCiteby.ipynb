```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import pandas as pd
```

```python
from selenium import webdriver

driver = webdriver.Chrome()
driver.set_window_size(1200, 800)
driver.get('https://www.scopus.com/')
#driver.get('https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&scp=85052201238&origin=inward')
```

```python
import re
def get_navigator(path):
    year_ele = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, path)))
    # print(view_more.text)
    try:
        view_more = year_ele.find_element(By.CSS_SELECTOR, 'ul[class="list-inline refineResultsViewLink"] > li')
        view_more.click()
    except Exception as e:
        print(".",end="")

    sub_year_ele = year_ele.find_elements(By.CSS_SELECTOR, 'li')
    res = []
    for sy in sub_year_ele:
        sy_split = sy.text.split()
        if len(sy_split) != 0:
            if re.match(r"^\(.*\)$", sy_split[-1]):
                topic = sy.find_element(By.CSS_SELECTOR,'span[class="btnText"]')
                # print(topic.text)
                count = sy.find_element(By.CSS_SELECTOR,'span[class="badge"] > span[class="btnText"]')
                # print(count.text)
                merge_par = f'''{topic.text}~{count.text}'''
                # print(merge_par)
                res.append(merge_par)

    return res

```

```python
df= pd.read_csv(r"C:\TESTCPP\Datasci\Final_project\target3.csv")
for i in range(0,df.shape[0]):
    filename = df['filename'][i]
    path = df['citedby_link'][i]
    driver.get(path)
    Cite_num = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.CSS_SELECTOR, 'h1 > span'))
    ).text
    #print(Cite_num)
    if(Cite_num == "0 document cited by results"): 
        output = [filename, 0, [], [], []]
        with open(r'C:\TESTCPP\Datasci\Final_project\Citebydata.csv', mode = 'a', newline='', encoding='utf-8') as csv_file:
            csv_writer = csv.writer(csv_file)
            csv_writer.writerow(output)
        print (output)
        continue
    year = get_navigator('div[id="navigator_2"]')
    subject = get_navigator('div[id="navigator_4"]')
    country = get_navigator('div[id="navigator_11"]')
    output = [filename, Cite_num, year , subject, country]
    with open(r'C:\TESTCPP\Datasci\Final_project\Citebydata.csv', 'a', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(output)
    print (output)
# print(get_navigator('div[id="navigator_2"]'))
# print(get_navigator('div[id="navigator_4"]'))
# print(get_navigator('div[id="navigator_11"]'))
```
