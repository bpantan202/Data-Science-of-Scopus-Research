```python
def next_page():
    next_button = driver.find_elements(By.CSS_SELECTOR, 'button[class="Button-module__nc6_8 Button-module__rphhF Button-module__VBKvn Button-module__MlsfC Button-module__hK_LA Button-module__qDdAl Button-module__rTQlw"]')
    next_button[-1].click()
```

```python
def get_one_page(output_file):
    WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'thead[class="TableHeader-module__R_KvD"]')))
    citations_dis = driver.find_elements(By.CSS_SELECTOR, 'tr[class="TableItems-module__m0Z0b TableItems-module__A6xTk"]')
    for ci in citations_dis:
        order_ele = ci.find_element(By.CSS_SELECTOR, 'span[class="Checkbox-module__AScP8"]')
        order = order_ele.text
        # print(order)
        title_ele = ci.find_element(By.CSS_SELECTOR, 'span[class="Typography-module__lVnit Typography-module__Nfgvc Button-module__Imdmt"]')
        title = title_ele.text
        # print(title)
        main_link_ele = ci.find_element(By.CSS_SELECTOR, 'a[class="Button-module__f8gtt Button-module__rphhF Button-module__VBKvn Button-module__ZS4lL Button-module__hK_LA Button-module__qDdAl Button-module__rTQlw"]')
        main_link = main_link_ele.get_attribute('href')
        # print(main_link)
        ele = ci.find_elements(By.CSS_SELECTOR, 'a[class="Button-module__f8gtt Button-module__rphhF Button-module__VBKvn Button-module__MlsfC Button-module__Y0far Button-module__hK_LA Button-module__qDdAl Button-module__rTQlw"]')
        amount = '0'
        href = None
        
        # if ele[-1].text.isdigit():
        #     amount = ele[-1].text
        #     href = ele[-1].get_attribute('href')
        
        # print(amount)
        # print(href)

        with open(output_file, mode='a', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            row = [order,amount,title,href,main_link]
            print("info =>",row)
            writer.writerow(row)


```

```python
import re

def get_navigator(path):
    print(".", end='')
    year_ele = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, path)))

    sub_year_ele = year_ele.find_elements(By.CSS_SELECTOR, 'li')
    res = []
    for sy in sub_year_ele:
        sy_split = sy.text.split()
        if len(sy_split) != 0:
            if re.match(r"^\(.*\)$", sy_split[-1]):
                topic = sy.find_element(By.CSS_SELECTOR,'span[class="btnText"]')
                # print(topic.text)
                count = sy.find_element(By.CSS_SELECTOR,'span[class="badge"] > span[class="btnText"]')
                # print(count.text)
                merge_par = f'''{topic.text}~{count.text}'''
                # print(merge_par)
                res.append(merge_par)
    return res
```

```python
import pandas as pd
import csv

df = pd.read_csv('KU_research_link.csv')

for i in range(370,1140):
    link = df['cities_link'][i]
    driver.get(link)

    WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'div[class="paddingTopOneHalf"]')))

    cited_ele = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CSS_SELECTOR, 'h1 > span')))
    cited_count = cited_ele.text

    year = get_navigator('div[id="navigator_2"]')
    subject = get_navigator('div[id="navigator_4"]')
    country = get_navigator('div[id="navigator_11"]')

    with open('bb_KU_cited.csv', mode='a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        row = [i+1, cited_count, year, subject, country]
        print("info =>",row)
        writer.writerow(row)
```
